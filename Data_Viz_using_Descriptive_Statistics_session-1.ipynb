# Descriptive Statistics:
### Why statistics?


Statistical analyses are used to answer many important questions in everyday life, such as the following



1. **Shopping Trends:** How can store managers predict which products will be popular during a sale to  ensure they stock the right items?

  Example: Store managers analyze past sales data to identify trends in product popularity. They calculate the mean and median sales for different products and use the standard deviation to predict future sales with a certain level of variability.

2. **Traffic Management:** How can city planners analyze traffic patterns to reduce congestion and improve commute times?

  Example: City planners collect data on traffic flow at different times of the day. They calculate the mean and median traffic volumes and the standard deviation to identify peak congestion times and develop strategies to mitigate traffic jams.


3. **Predicting Weather:** How can meteorologists predict tomorrow’s weather to help people plan their day?

  Example: Meteorologists use historical weather data, including temperature, humidity, and wind speed, to predict the weather. They analyze patterns and calculate the mean, median, and standard deviation to estimate tomorrow’s weather within a range of variability.

4. **Customer Satisfaction:** How can companies analyze feedback to improve their products and services, ensuring customer satisfaction?

  Example: Companies gather customer feedback and satisfaction scores. They calculate the mean and median satisfaction levels and the standard deviation to understand the variability in customer experiences and identify areas for improvement.


4. **Sports Performance:** How can coaches analyze players’ performance data to make strategic decisions and improve team success?

  Example: Coaches track players' performance metrics, such as points scored and assists. They calculate the mean and median performance levels and the standard deviation to identify consistent players and areas for improvement.

5. **Sports Performance:** How can factories ensure the accuracy and consistency of their product quantities?

  Example: In a factory, every hour, a technician randomly selects 30 cool drink bottles and measures their liquid content. They calculate the mean, median, mode, range, variance, and standard deviation to ensure the bottles contain close to 500 ml. This ensures the production process is accurate and consistent.


Similarly,

* How can an investor predict the behavior of stock markets to make profitable investments?

* How can a real-estate agent judge in which area to buy houses so that they get the best quality for the lowest prices?

### Quartiles:

Previously, you learnt about quantiles. Quartiles are measures of position that divide a data set into four equal parts.
* What if we wanted to divide the compensation values into four equal parts?
image.png

* These three values are known as quartiles.
* The first or the lower quartile is ₹7,49,836, the second quartile (median) is 8,51,241, and the third or the upper quartile is 10,98,774.
* Using quartiles instead of the median also gives us a sense of the distribution of the values.


The three quartiles are given below:
1. First or lower quartile
2. Second quartile or the median
3. Third or upper quartile

Though the zeroth and fourth quartiles are not universally accepted, one may consider the zeroth quartile as the lowest value of the data and the fourth quartile as the highest value of the data. However, we will consider the convention that three quartiles are available and discuss them.

The first quartile is the value below which 25% of the data lies.
This is referred to as the exclusive definition of quartiles (and quantiles in general). The inclusive definition of quantiles includes the quantile value as well.
In that sense, 25% of the data would be at or below the first quartile. Either definition of quantiles can be used based on the situation.

Next, let’s consider an example of the employee specifics of a company to understand quartiles better. Suppose the data of the compensation of the employees is divided into four equal parts. Let’s take a look at the histogram depicting the quartiles.



image.png

You can observe that three red lines divide the data into four equal parts. Each part contains the same number of data points. The red lines represent the first, the second and the third quartiles.

Now, you will learn about the technique to find the quartiles. Consider the data of the age of 21 students in a school that is given below.
image.png

Here are the steps to follow to find the quartiles of a data set:
* Step 1: Arrange the data set in the ascending or descending order.
* Step 2: Find the median of the entire data set, which is the second quartile.
* Step 3: Consider the data set on each side of the median. One from the left and another from the right of the median of the entire data set.
* Step 4: Find the median of the data sets extracted from both sides of the median of the entire data set.


First, we need to arrange the age in ascending or descending order. After arranging the data in ascending order, we get 11, 11, 12, 12, 13, 13, 13, 14, 14, 14, 15, 15, 16, 16, 16, 17, 17, 17, 18, 18, 19. So, the median is 15, which is in the 11th position. The median (or the second quartile) divides the data into two equal parts.

Now, the first 10 ages that are on the left side of the median are 11, 11, 12, 12, 13, 13, 13, 14, 14, 14. Now, the median of these 10 data points is (13+13)2=13. This is the first quartile.

Similarly, the median of the last 10 ages that are on the right side of the median of the entire data set is 17+172=17, which is the third quartile.

Thus, we can find all the three quartiles of the given data set as follows:
* First quartile: Generally represented by Q1=13
* Second quartile: Generally represented by Q2=15
* Third quartile: Generally represented by Q3=17


**Interquartile range (IQR):** In a data set, we can measure the three quartiles Q1, Q2 and Q3. The value Q3−Q1 is called the interquartile range. In this example, the IQR is 17−13=4. The IQR is a measure of the difference between the lower and the upper bounds of data that contain 50% of the data. It essentially gives us the range of normalcy of a data set.
So, you can conclude that the first 25% students are aged less than 13 years, the first 50% students are aged less than 15 and the first 75% students are aged less than 17 years, and the IQR is 4 years. This is how you can use quartiles to analyze a data set.

There are other measures of variability that do better at capturing the variation in data, such as mean absolute deviation and standard deviation. Before moving on, please attempt the following questions.


**​MEAN ABSOLUTE DEVIATION**

Deviation indicates how much a particular data point varies from a fixed value. More often than not, this fixed value is usually selected as some measure of central tendency. So, to find the mean absolute deviation, we first need to find all the deviations from a fixed value such as the mean or the median and then take the mean of all the deviations. Let’s hear what our subject matter expert, Anjali has to say about this.

Mean absolute deviation about the mean:
The absolute values of the deviations of each data point from the mean of a data set is called the absolute deviation about the mean. The mean of all the absolute deviations is called the mean absolute deviation about the mean. This helps to measure the variability of a data set.

Let’s take an example and follow a step-by-step process to find the mean absolute deviation about the mean.

Consider the employee specifics of two companies, Company A and Company B, given below.

image.png


Here are the steps to calculate the mean absolute deviation about the mean for Company A:

Step 1: Find the mean of the compensation. As you can see, the compensation details of 10 employees are given. Here, the mean of the data is ₹8,14,272.40.

Step 2: Find the absolute deviations of each data point from the mean. For example, the absolute deviation of the data point ₹8,65,212 from the mean ₹8,14,272.40 is equal to |8,65,212−8,14,272.40|=50939.60 and for the data point ₹7,34,712, it is equal to |7,34,712−8,14,272.40|=79560.40. Similarly, find the absolute deviations for all the data points.

Step 3: Find the mean of the deviations that you have calculated in Step 2. For this given data set, the mean absolute deviation about the mean is , which is equal to ₹1,22,758.28.

Similarly, you can find the mean absolute deviation about the mean for Company B, which comes out to be ₹2,56,225.50.

Note that the mean of compensation values for Company A is ₹8,14,272.40 and that for Company B is ₹10,14,440.30. And, their respective mean absolute deviations are ₹1,22,758.28 and ₹2,56,225.50, which differ by about ₹1,00,000. This basically indicates that the variation in compensation values for Company B is apparently more than that for Company A.

Let’s formalize your learnings. If x1, x2, x3, … , xn are n data points with mean ¯X, then the mean absolute deviation about the mean is ( |x1−¯X|+|x2−¯X|+|x3−¯X|+...+|xn−¯X| ) / n=(∑ni=1|xi−¯X| ) / n.

The mean absolute deviation about the median is a similar concept. Here, you need to find the absolute deviation from the median and the remaining process is the same.

Here are the steps to calculate the mean absolute deviation from the median:
* Step 1: Calculate the median of the data set.
* Step 2: Calculate the absolute values of the differences between the values in the data set and the median.
* Step 3: Calculate the mean of these absolute differences.

To formalize this, if x1, x2, x3, … , xn are n data points with median X, then the absolute mean deviation about the median is (|x1−X|+|x2−X|+|x3−X|+...+|xn−X|) / n=(∑ni=1|xi−X|) / n.

​So far, we have understood the concept of mean absolute deviation from the mean as well as the median, but in most cases, we will use the mean absolute deviation about the mean. This is because the mean of a data set is calculated by considering the entire data, including all the outliers as well. This gives us a better sense of the measure of dispersion in the data set.


Next, we will study the variance and the standard deviation of a data set.

1. The variance of the values in a data set is similar to the mean absolute deviation about the mean, but here, the deviations are squared and added together
2. To find the variance, we use the following algorithm:
  * Calculate the mean of the data set
  * Calculate the values of the differences between the values in the data set and the mean
  * Calculate the square of these deviation values
  * Calculate the mean of the squares of the deviations
